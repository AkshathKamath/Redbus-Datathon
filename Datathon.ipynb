### Comparing std dev for SEAT TYPE 1

```python
Bus=price_df['Bus'].unique()
len(Bus)
```

```python
Dates=price_df['Service Date'].unique()
len(Dates)
```

```python
std_dev_comp=pd.DataFrame() #1
Bus_add=[]
Date_add=[]
std_add=[]
```

```python
for bus in Bus: #2
    for date in Dates:
        std=price_df[(price_df['Bus']==bus)&(price_df['Service Date']==date)]['Seat Fare Type 1 Mean'].std()
        Bus_add.append(bus)
        Date_add.append(date)
        std_add.append(std)
```

```python
std_dev_comp.shape
```

```python
std_dev_comp.info()
```

```python
std_dev_comp['Bus']=Bus_add
std_dev_comp['Service Date']=Date_add #3
std_dev_comp['std']=std_add
```

```python
std_dev_comp[std_dev_comp['Bus']=='d6fa79179fda2a77455794637f225962'].shape[0]
```

```python
std_dev_comp.head(20)
```

```python
std_dev_comp['std']=std_dev_comp['std'].replace(0.0,np.nan) #4
```

```python
std_dev_comp.dropna(axis=0,how='any',inplace=True)
```

```python
std_dev_comp[std_dev_comp['Service Date']=='20-07-2020 00:00'].head(117)
```

```python
sample=pd.DataFrame()
```

```python
score=pd.DataFrame()
score['Bus']=Bus2
score['score']=0.0
score.head()
#score.shape
```

```python
for date in Dates:
    sample=std_dev_comp[std_dev_comp['Service Date']==date].sort_values('std',axis=0,ascending=True)
    rows=sample.shape[0]
    df_1=sample.iloc[:int(rows/3),:] # Dividing arranged df into 3 & giving score of 10 to 1st 1/3rd with lowest std dev & corr.
    df_2=sample.iloc[int(rows/3):int(2*rows/3),:]
    df_3=sample.iloc[int(2*rows/3):,:]
    for bus in df_1['Bus']:
        score.loc[score['Bus']==bus,'score']+=10.0
    for bus in df_2['Bus']:
        score.loc[score['Bus']==bus,'score']+=9.0
    for bus in df_3['Bus']:
        score.loc[score['Bus']==bus,'score']+=8.0
```
